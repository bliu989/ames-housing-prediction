# Importing data

The data used here is from a [Kaggle dataset](

```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Lasso

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score

from sklearn.feature_selection import RFE

from sklearn.metrics import mean_squared_error
```

```python
# read data

df = pd.read_csv("AmesHousing.csv")
df.head()
```

# Cleaning data

```python
# show which columns have missing values

null_counts = df.isnull().sum()
null_counts[null_counts != 0]
```

```python
# drop columns with too many (> 200) missing values

df = df.drop(null_counts[null_counts > 200].index, axis = 1)
df.head()
```

```python
# drop irrevalent columns

df = df.drop(['Order', 'PID'], axis = 1)
df.head()
```

```python
# visualize where the missing values are

plt.figure(figsize = (10, 20))
ax = sb.heatmap(df.isnull())
```

```python
# drop entries with missing values

df = df.dropna()
df.head()
```

```python
# set MS SubClass column as categorical

df = df.astype({'MS SubClass' : 'object'})
```

```python
col_dtypes = df.dtypes
categorical_columns = col_dtypes[col_dtypes == 'object'].index

# Perform one-hot encoding on categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns)
```

```python
df_encoded.head()
```

```python
# see which levels of the categorical variables cover a few or almost all entries
counts = df_encoded[df_encoded.dtypes[df_encoded.dtypes == 'bool'].index].sum()
plt.hist(counts, bins = 20)
plt.xlabel('Number of entries with a particular level')
plt.ylabel('Number of levels')
plt.show()
```

```python
# drop the unrepresentative levels
unrep_levels = df_encoded[counts[(counts < 250) | (counts > 2500)].index]

df_clean = df_encoded.drop(unrep_levels, axis = 1)
df_clean.head()
```

```python
# split data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_clean.drop('SalePrice', axis=1), 
                                                    df_clean['SalePrice'], 
                                                    test_size=0.2, 
                                                    random_state=20)
```

```python
# taking the log of the SalePrice
y_train = np.log(y_train)
y_test = np.log(y_test)
```

# Exploring the data

```python
with pd.option_context('display.max_columns', None):
    display(X_train.describe())

y_train.describe()
```

```python
non_bool_cols = X_train.dtypes[X_train.dtypes != 'bool'].index

# calculate the skewness of the quantitative features
skewness = X_train[non_bool_cols].skew()

plt.figure(figsize=(8,8))
skewness.sort_values().plot.barh()
plt.xticks(np.arange(0,21,1))
plt.vlines(1, -5, 40, colors = 'red')
plt.xlabel('Skewness')
plt.show()
```

```python
# attempting to correct the skewness with a logarithm
skew_cols = non_bool_cols[skewness > 1]
for col in skew_cols:
    X_train[col] = np.log1p(X_train[col])
```

```python
# calculate skewness again
skewness = X_train[non_bool_cols].skew()

plt.figure(figsize=(8,8))
skewness.sort_values().plot.barh()
plt.xticks(np.arange(0,21,1))
plt.vlines(1, -5, 40, colors = 'red')
plt.xlabel('Skewness')
plt.show()
```

```python
# look at the distribution of the variables that are still skewed
# using their values before the logarithm
still_skewed = X_train[non_bool_cols[skewness > 1]].columns

fig, axes = plt.subplots(3, len(still_skewed) // 3, figsize=(12, 10))

for i, column in enumerate(still_skewed):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    
    ax.hist(np.expm1(X_train[column]))
    ax.set_title(column)
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```

```python
(X_train[non_bool_cols[skewness > 1]] > 0).sum()
```

```python
(X_train['Kitchen AbvGr'] > 1).sum()
```

These features are very skewed and have few values above the minimum value, so I'll drop these features.

```python
X_train = X_train.drop(still_skewed, axis = 1)
```

```python
# check for linear relationships with SalePrice
X_train_num = X_train.select_dtypes(exclude='bool')

fig, axes = plt.subplots(5, 5, figsize=(14, 14))
for i, column in enumerate(X_train_num.columns):
    row = i // 5
    col = i % 5
    ax = axes[row, col]
    
    ax.scatter(X_train_num[column], y_train)
    ax.set_xlabel(column)
    ax.set_ylabel("Sale Price")

plt.tight_layout()
plt.show()
```

```python
# performing the same transformations on X_test
for col in skew_cols:
    X_test[col] = np.log1p(X_test[col])

X_test = X_test.drop(still_skewed, axis = 1)
```

# Fitting a linear regression

I'll try to fit a linear regression model with only 5 predictors using different techniques, since this gives the model better interpretability. Then I'll fit a linear regression model with all the predictors. 

## Selecting features using RFE

```python
model = LinearRegression()
num_features_to_select = 5

# Create an RFE selector 
rfe = RFE(estimator=model, n_features_to_select=num_features_to_select)
rfe.fit(X_train, y_train)

# Retrieve the selected features
selected_features = X_train.columns[rfe.support_]

print("Selected Features:", selected_features)
```

```python
model.fit(X_train[selected_features], y_train)
```

```python
# get coefficients for the linear model
[f'{x:.8f}' for x in model.coef_]
```

## Manually selecting features

```python
# get features most correlated with SalePrice
corrs = X_train.corrwith(y_train)
corrs_sorted = corrs.sort_values(key = abs, ascending=False)
plt.barh(corrs_sorted.index[:20], corrs_sorted[:20])
plt.show()
```

```python
model_manual = LinearRegression()
```

```python
# manually pick 5 features, trying to avoid collinearity
manual_selected_features = ['Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Full Bath', 'Year Built']
model_manual.fit(X_train[manual_selected_features], y_train)
```

```python
# get coefficients for the linear model
[f'{x:.8f}' for x in model_manual.coef_]
```

What's strange here is that 'Full Bath' is given a negative coefficient.

## Selecting features using Lasso

```python
# finding a good value for the alpha parameter for Lasso

alphas = np.logspace(-3, 3, 7)  # Alpha values ranging from 10^-3 to 10^3

mean_scores = []
nonzeros = []

for alpha in alphas:
    lasso = Lasso(alpha=alpha)

    scores = cross_val_score(lasso, X_train, y_train, cv=5)  
    mean_score = np.mean(scores)
    mean_scores.append(mean_score)
    
    lasso.fit(X_train, y_train)
    num_nonzero_coeff = (~(lasso.coef_ == 0)).sum()
    nonzeros.append(num_nonzero_coeff)
```

```python
# summarize the alpha search
lasso_df = pd.DataFrame({'alpha': alphas, 
                         'Mean cross-validation scores': mean_scores, 
                         '# of nonzero coefficients': nonzeros})
lasso_df
```

```python
# search for alpha between 0.1 and 10
alphas = np.logspace(-1, 1, 10) 

mean_scores = []
nonzeros = []

for alpha in alphas:
    lasso = Lasso(alpha=alpha)

    scores = cross_val_score(lasso, X_train, y_train, cv=5)  
    mean_score = np.mean(scores)
    mean_scores.append(mean_score)
    
    lasso.fit(X_train, y_train)
    num_nonzero_coeff = (lasso.coef_ != 0).sum()
    nonzeros.append(num_nonzero_coeff)

# summarize the alpha search
lasso_df = pd.DataFrame({'alpha': alphas, 
                         'Mean cross-validation scores': mean_scores, 
                         '# of nonzero coefficients': nonzeros})
lasso_df
```

```python
# train Lasso with alpha = 0.278256
model_lasso = Lasso(alpha = 0.278256)

model_lasso.fit(X_train, y_train)
X_train.columns[model_lasso.coef_ != 0]
```

```python
# get coefficients for the linear model
[f'{x:.8f}' for x in model_lasso.coef_ if x != 0]
```

## Selecting all features

```python
model_all = LinearRegression()

model_all.fit(X_train, y_train)

# get RMSE on test set
y_pred = model_all.predict(X_test)

test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)

print(f"RMSE: {test_rmse:.4f}")
```

```python
# get coefficients for the linear model
print(*[f'{x:.8f}' for x in model_all.coef_], sep=', ')
```

# Random forest regression

Since SalePrice does not necessarily have a strong linear relationship with some features as discovered in the exploratory process, I'll also build a model using a random forest. However, this loses interpretability compared to a linear regression with 5 predictors.

```python
rf_model = RandomForestRegressor(random_state = 20)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
```

```python
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, 
                           cv=5, 
                           scoring='neg_mean_squared_error', 
                           verbose=3)
grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Mean Squared Error:", -grid_search.best_score_)
```

```python
best_rf_model = RandomForestRegressor(random_state = 20,
                                     max_depth = 30,
                                     min_samples_split= 5,
                                     n_estimators = 200)
best_rf_model.fit(X_train, y_train)
```

# Gradient boosting

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3]
}
```

```python
gbm = GradientBoostingRegressor(random_state=20)

grid_search = GridSearchCV(estimator=gbm, 
                           param_grid=param_grid, 
                           cv=5, 
                           scoring='neg_mean_squared_error',
                           verbose=3)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_mse = -grid_search.best_score_
print(f"Best Hyperparameters: {best_params}")
print(f"Best Mean Squared Error: {best_mse}")
```

```python
best_gb_model = GradientBoostingRegressor(random_state=20,
                                         learning_rate=0.1,
                                         max_depth=3,
                                         min_samples_leaf=3,
                                         min_samples_split=2,
                                         n_estimators=200)
best_gb_model.fit(X_train, y_train)
```

# Evaluating all the models

```python
# calculate the RSME for all models

all_features = X_test.columns

models = [('Linear with RFE', model, selected_features),
          ('Linear with manual feature selection', model_manual, manual_selected_features),
          ('Linear with Lasso', model_lasso, all_features),
          ('Linear with all predictors', model_all, all_features),
          ('Random forest', best_rf_model, all_features),
          ('Gradient boosting', best_gb_model, all_features)]

for model_name, mod, features in models:
    y_pred = mod.predict(X_test[features])

    test_mse = mean_squared_error(y_test, y_pred)
    test_rmse = np.sqrt(test_mse)

    print(f'{model_name} RMSE: {test_rmse:.4f}')
```
